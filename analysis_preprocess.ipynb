### generate target list which is for dataloader
target list is a list of (chid, dt, shop_tag_list, txn_amt_list)

```python
import os
import pickle
from tqdm import tqdm
from multiprocessing import Pool, Manager

manager = Manager()
target_list = manager.list()
customer_dir = 'data/raw/raw_split_by_customer'

def get_target(path):
    df = pd.read_csv(f"{customer_dir}/{path}", usecols=['dt', 'shop_tag', 'txn_amt'])
    dts = df.dt.unique()
    # drop first month
    if len(dts) == 1: return
    dts = dts[1:]
    # preprocess  
    df['shop_tag'] = df['shop_tag'].apply(lambda x: 49 if x == 'other' else int(x)).astype(int)
    df['shop_tag'] = df['shop_tag'] - 1
    
    chid = int(path.split('.')[0])
    for dt in dts:
        dft = df[df['dt'] == dt]
        target = [chid, dt, dft['shop_tag'].tolist(), dft['txn_amt'].tolist()]
        target_list.append(target)

with Pool(16) as p:
    paths = os.listdir(customer_dir)
    with tqdm(total=len(paths)) as pbar:
        for i, _ in enumerate(p.imap_unordered(get_target, paths)):
            pbar.update()
```

### concat pkl for data loader

```python
import os
import pandas as pd
import pickle
from tqdm import tqdm
from multiprocessing import Pool, Manager

manager = Manager()
x_dict = manager.dict()
raw_customer_dir = '/media/hd03/axot_data/cc/data/raw/raw_split_by_customer'
customer_dir = '/media/hd03/axot_data/cc/data/preprocessed/v1/customers_pkl'

def concat(chid):
    raw_df = pd.read_csv(f"{raw_customer_dir}/{chid}.csv", usecols=['dt', 'txn_amt'])
    df = pd.read_pickle(f"{customer_dir}/{chid}.pkl")
    df['txn_amt_pct'] = raw_df.groupby('dt').transform(lambda x: x/x.sum())['txn_amt']
    x_dict[chid] = df.values

chids = list(range(10_000_000, 10_000_000+500_000))
with Pool(16) as p:
    with tqdm(total=len(chids)) as pbar:
        for i, _ in enumerate(p.imap_unordered(concat, chids)):
            pbar.update()

x_dict = dict(x_dict)
print(f'len x_dict: {len(x_dict)}')
pickle.dump(x_dict, open('/media/hd03/axot_data/cc/data/preprocessed/v1/customers_txn_amt_pct.pkl', 'wb'))
```

## model
for each customer
- for each month
    - for each row of a month
        - run row encoder to get emb which takes row data as input
    - padding and truncate to fixed size i.e.10 rows 
    - aggregate rows information by rows aggregator i.e (self-attention + pooling), (conv) to get month embedding.
- padding to fixed size i.e. 12 months
- run temporal information aggregator module i.e. rnn-based, conv and return 49 probalilities.
